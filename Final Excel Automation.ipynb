# Removing only Images in Excel files

```python
import xlwings as xw
import glob
import os


# Give Your Location where excel file are thare  <--------------------
path0 = "C:\\Users\\bp_Utkarsh Tripathi\\Desktop\\1A Jupyter notebook"
path = path0.replace("\\", "//") + "//"
print(path)
print( f" Total --->", len(glob.glob(path + "*.xlsx")),  "Excel files hai \n")


for i, j in enumerate(glob.glob(path + "*.xlsx")):
    
    only_file_name = os.path.basename(j).split('/')[-1]
    
    images = xw.Book(j).sheets.active.pictures   #<----- getting images
    
    for image in images:                         #<----- removing images
        image.delete()

    xw.Book(j).save()
    print(f"{i+1} ------> {only_file_name} ✅ Done \n\n")

    xw.apps.active.quit()                       #<---- for closing Excel windows
    
    
print("\n\n Everything done Just Chill.....👍😊 \n\n ")
```
